# Data Visualization


```python
import pandas as pd 
%matplotlib inline
import matplotlib.pyplot as plt
```

## Data contains:

- TV promotion budget (in million)
- Social Media promotion budget (in million)
- Radio promotion budget (in million)
- Influencer: Whether the promotion collaborate with Mega, Macro, Nano, Micro influencer
Sales (in million)

## Load the data 

```python
stock_prices = pd.read_csv("Socialbudget.csv")
stock_prices.head()
```

```python
stock_prices.describe()
```

```python
stock_prices.info()
```

```python
stock_prices.corr()
```

```python
import seaborn as sns
```

## rate of influencers in the dataset

```python

sns.set()
sns.countplot(data=stock_prices, x='Influencer')
```

## TV promotion budget depending on influencers

```python
stock_prices = stock_prices.iloc[:500]
TV_ = sns.barplot(data=stock_prices, x="Influencer", y="TV")
```

```python
sns.relplot(data=stock_prices, x="Influencer", y="TV", hue="Sales")
```

## TV sales depending on influencers sales

```python
sns.lmplot(x="Sales", y="TV", data=stock_prices)
```

## Radio promotion budget depending on influencers

```python
Radio_ = sns.barplot(data=stock_prices, x="Influencer", y="Radio")
```

```python
sns.relplot(data=stock_prices, x="Influencer", y="Radio", hue="Sales")
```

## Radio sales depending on influencers sales

```python
sns.lmplot(x="Sales", y="Radio", data=stock_prices)
```

## Social Media promotion budget depending on influencers

```python
Social_Media_ = sns.barplot(data=stock_prices, x="Influencer", y="Social Media")
```

```python
sns.relplot(data=stock_prices, x="Influencer", y="Social Media", hue="Sales")
```

## Social Media sales depending on influencers sales

```python
sns.lmplot(x="Sales", y="Social Media", data=stock_prices)
```
